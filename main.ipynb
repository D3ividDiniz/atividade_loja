# 5 - AOV X Preço Médio de Venda

###### Calcule o AOV e o Preço Médio de Venda mês a mês ao longo do período mensurado, crie também uma nova coluna "AOV_PMV" que indicará a razão entre AOV/PMV. Analise a variação dessa nova coluna mês a mês, e interprete o resultado com base na comparação entre AOV e PMV.

# 8 - Retention e Churn Rate




##### A taxa de retenção (retention rate) refere-se a razão entre os clientes que continuam a comprar os produtos de um negócio em um determinado período de tempo e o total de clientes. A taxa churn refere-se ao oposto, ou seja, a razão entre os clientes que deixaram de comprar os produtos de um negócio em um determinado período de tempo e o total de clientes. Um bom negócio possui uma alta taxa de retenção e uma baixa taxa churn.




###### **<span style = 'color:red'>OBS: Considere a utilização de nunique() para a contagem de elementos únicos do DataFrame</span>**




```python
df_8 = df

df_8 = df_8.groupby(["Year","Customer ID"]).agg({"Order ID":"count"})

anos = []
cont = 0

for ano in df.groupby("Year").mean().index:
  lastyear = df_8.query(f"Year == {ano}")
  ids = []
  for year in lastyear.index:
    ids.append(year[1])
  
  anos.append(ids)

response = [];
others = []
for index,listYear in enumerate(anos[0:3]):
  number = 0 
  other = 0 
  for cu in listYear:
    if cu in anos[index+1]:
      number+=1
    else:
      other +=1
  response.append(number)
  others.append(other)

years = df.groupby("Year").mean().index.tolist()   
for index,ano in enumerate(anos[1:]):
  print(f"{years[index]} -- {years[index +1]} ")
  chorn = round(response[index]/(others[index]+response[index]),2)
  print(str(chorn)+" C  --  "+str(round(1-chorn,2))+" R")
```
